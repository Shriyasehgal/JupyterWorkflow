```python
#numpy- adds multidimention support to visual data
#matplotlib- represnation
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
data = pd.read_csv('data.csv')
gender = {'M': 0,'F': 1}
data.Gender = [gender[item] for item in data.Gender] 
data
```

```python
dfx=data.iloc[:,0:5]
dfx
```

```python
dfy=data.iloc[:,5]
dfy
```

```python
x_train,x_test,y_train,y_test=train_test_split(dfx,dfy,test_size=0.25)
```

```python
#decision tree
dt = DecisionTreeClassifier()
#fitting th data
dt.fit(x_train,y_train)


dt.score(x_test,y_test)
```

```python
dt.score(x_train,y_train) #this is overfitting.. therefore predcition is not very good
```

```python
#HOW THE SCORE WAS ACTUALLY CALCULATED
pred = dt.predict(x_test)
pred
```

```python
actual=np.array(y_test)
actual
```

```python
count=0
for i in range (len(pred)):
    if pred[i]==actual[i]:
        count=count+1
        
count
```

```python
len(pred)
```

```python
acuracy= count/len(pred)
acuracy
```

```python
#implementing Random Forest
rf=RandomForestClassifier(n_estimators=10)
rf.fit(x_train,y_train.values.ravel())
#.values will give the values in an array. (shape: (n,1)
#.ravel will convert that array shape to (n, )
#The problem was that the labels were in a column format while it expected it in a row. use np.ravel()
```

```python
rf.score(x_test,y_test)
```

```python
rf.score(x_train,y_train) # not completely overfitting
```

```python
#HOW SCORE WAS ACTUALLY CALCULATED
pred = rf.predict(x_test)
pred
```

```python
actual=np.array(y_test)
actual
```

```python
count=0
for i in range (len(pred)):
    if pred[i]==actual[i]:
        count=count+1
        
count
```

```python
accuracy= count/ len(pred) 
accuracy
```

```python
#BAGGING CLASSIFIER
bg=BaggingClassifier(DecisionTreeClassifier(), max_samples=0.5, max_features=1.0)
bg.fit(x_train,y_train.values.ravel())
```

```python
bg.score(x_test,y_test)
```

```python
bg.score(x_train,y_train)
```

```python
#boosting reduces overfitting and variance
adb = AdaBoostClassifier(DecisionTreeClassifier(),learning_rate=1)
```

```python
adb.fit(x_train,y_train.values.ravel())
adb.score(x_test,y_test)
```

```python
#USING 3 CLASSIFIERS TOGETHER
lr = LogisticRegression()
dt=DecisionTreeClassifier()
svm=SVC(kernel='poly',degree=2, gamma='auto')
```

```python
evc = VotingClassifier( estimators=[('lr',lr),('dt',dt),('svm',svm)],voting='hard')
evc.fit(x_train,y_train.values.ravel())
```

```python
evc.score(x_test,y_test)
```
